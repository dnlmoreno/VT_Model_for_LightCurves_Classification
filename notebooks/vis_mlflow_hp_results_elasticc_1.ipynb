```python
import mlflow
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
mlflow.set_tracking_uri("file:./results/elasticc_1/ml-runs")
```

```python
data_name =  'elasticc_1'
experiment_name = f"ft_classification/{data_name}/testing"

experiment = mlflow.get_experiment_by_name(experiment_name)
if experiment:
    experiment_id = experiment.experiment_id
    print(f"Experiment ID for '{experiment_name}': {experiment_id}")
else:
    print(f"Experiment '{experiment_name}' not found.")

runs = mlflow.search_runs(experiment_ids=[experiment_id])

# Filtrar las corridas que coincidan con los par√°metros especificados
filtered_runs = runs[
    (runs['params.imgs_params/input_type'] == 'overlay')
]

failed = filtered_runs[filtered_runs.status == 'FAILED']
running = filtered_runs[filtered_runs.status == 'RUNNING']
finished = filtered_runs[filtered_runs.status == 'FINISHED']

print(len(failed))
print(len(running))
print(len(finished))

grouped_df = finished.groupby([
    'params.imgs_params/fig_params/linewidth',
    'params.imgs_params/fig_params/markersize',
    'params.imgs_params/input_type',
    'params.imgs_params/use_err',
    'params.training/lr',
])['metrics.f1/test'].agg(['mean', 'std']).reset_index()

grouped_df.rename(columns={'mean': 'metrics.f1/test_mean', 'std': 'metrics.f1/test_std'}, inplace=True)

print(f'Number experiments: {len(grouped_df)}')

grouped_df = grouped_df.sort_values(by=['metrics.f1/test_mean'], ascending=False)
grouped_df.head(10)
```
