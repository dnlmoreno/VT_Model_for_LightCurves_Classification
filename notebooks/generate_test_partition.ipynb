```python
import pandas as pd
```

```python
path_data = 'data/lightcurves/plasticc/raw'
target_name = 'true_target'
df_metadata_test = pd.read_csv(f'{path_data}/unblinded_test_set_metadata.csv')
df_metadata_test = df_metadata_test[~df_metadata_test[target_name].isin([991, 992, 993, 994])]
df_metadata_test
```

```python
df_metadata_test.groupby('true_target').count()['object_id']
```

```python
test_2M = pd.read_parquet('data/lightcurves/plasticc/partitions/test/test_0.6/test.parquet')
test_2M
```

```python
df_obj_label_test = df_metadata_test[df_metadata_test.object_id.isin(test_2M.object_id)][['object_id', 'true_target']].reset_index(drop=True)
df_obj_label_train_val = df_metadata_test[~df_metadata_test.object_id.isin(test_2M.object_id)][['object_id', 'true_target']].reset_index(drop=True)
df_obj_label_train_val
```

```python
df_obj_label_test
```

```python
desired_counts = {
    6: 151,
    15: 495,
    16: 924,
    42: 1193,
    52: 183,
    53: 30,
    62: 484,
    64: 102,
    65: 981,
    67: 208,
    88: 370,
    90: 2313,
    92: 239,
    95: 175
}
desired_counts
```

```python
filtered_df = pd.DataFrame()

# Iterar sobre cada grupo en el DataFrame
for target, count in desired_counts.items():
    # Filtrar el DataFrame para incluir solo las filas del label actual
    subset = df_obj_label_train_val[df_obj_label_train_val['true_target'] == target]
    
    # Comprobar si el DataFrame filtrado tiene más filas que el conteo deseado
    if len(subset) >= count:
        # Seleccionar aleatoriamente 'count' filas del subset
        sample = subset.sample(n=count, random_state=1)  # Usar random_state para reproducibilidad
    else:
        # Si no hay suficientes filas, toma todas las disponibles
        available_samples = subset.copy()
        remaining_count = count - len(available_samples)
        additional_samples = subset.sample(n=remaining_count, replace=True, random_state=1)
        sample = pd.concat([available_samples, additional_samples])

    # Agregar las filas seleccionadas al DataFrame final
    filtered_df = pd.concat([filtered_df, sample])

# Reiniciar el índice del DataFrame final
filtered_df.reset_index(drop=True, inplace=True)

# Mostrar o guardar el DataFrame resultante
filtered_df.groupby('true_target').count()
```

```python
##### AQUI TENGO QUE DECIDIR CUANTO OCUPAR
df_obj_aux = df_obj_label_train_val.copy()
#df_obj_aux = filtered_df.copy()

path_save_split = f'data/lightcurves/plasticc/partitions/test/stratified_5fold_splits_{df_obj_aux.shape[0]}_objs.parquet'
path_save_split
```

```python
from sklearn.model_selection import StratifiedKFold
import os

# Crear la carpeta si no existe
os.makedirs(os.path.dirname(path_save_split), exist_ok=True)

# Verificar si el archivo existe
if os.path.exists(path_save_split):
    # Abrir el archivo
    df_partitions = pd.read_parquet(path_save_split)
    print("Archivo de partición cargado correctamente.")
else:
    # Crear el objeto StratifiedKFold
    skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    # Particionar el dataset
    df_partitions = []
    for fold, (train_idx, val_idx) in enumerate(skf.split(df_obj_aux['object_id'], df_obj_aux['true_target'])):
        train_data = df_obj_aux.loc[train_idx]
        val_data = df_obj_aux.loc[val_idx]
        train_data['subset'] = f'train_{fold}'
        val_data['subset'] = f'val_{fold}'
        df_partitions.append(pd.concat([val_data, train_data]))

    df_partitions = pd.concat(df_partitions).reset_index(drop=True)
    df_obj_label_test['subset'] = 'test'
    df_partitions = pd.concat([df_partitions, df_obj_label_test]).reset_index(drop=True)
    df_partitions.to_parquet(f'{path_save_split}')

df_partitions
```
