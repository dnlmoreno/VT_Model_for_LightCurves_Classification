```python
import mlflow
import pandas as pd
import matplotlib.pyplot as plt
import os
#os.environ["PATH"] = "/home/users/dmoreno2016/.TinyTeX/bin/x86_64-linux:" + os.environ["PATH"]

plt.rcParams.update({
    "text.usetex": True,
    "pgf.texsystem": "pdflatex",
    "font.family": "serif",
    "font.serif": ["Palatino"],
})


import warnings
warnings.filterwarnings('ignore')
```

```python
mlflow.set_tracking_uri("file:./results/macho/ml-runs")
```

```python
dict_results = {
    'single_band': {
        'list_mean_by_spc': [],
        'list_std_by_spc': [],
    },
    'two_bands_grid':{
        'list_mean_by_spc': [],
        'list_std_by_spc': [],
    },
    'two_bands_overlay':{
        'list_mean_by_spc': [],
        'list_std_by_spc': [],
    }
}

def update_dict_results(df, dict_results, key, input_type):
    dict_results[key]['list_mean_by_spc'].append(
        grouped_df[grouped_df['params.imgs_params/input_type'] == input_type]['metrics.f1/test_mean'].iloc[0]
        )
    dict_results[key]['list_std_by_spc'].append(
        grouped_df[grouped_df['params.imgs_params/input_type'] == input_type]['metrics.f1/test_std'].iloc[0]
        )
    
    return dict_results
```

```python
data_name =  'macho'
experiment_name = f"ft_classification/{data_name}/testing"

experiment = mlflow.get_experiment_by_name(experiment_name)
if experiment:
    experiment_id = experiment.experiment_id
    print(f"Experiment ID for '{experiment_name}': {experiment_id}")
else:
    print(f"Experiment '{experiment_name}' not found.")

runs = mlflow.search_runs(experiment_ids=[experiment_id])

# Filtrar las corridas que coincidan con los parámetros especificados
filtered_runs = runs[
    (runs['params.loader/spc'] == '20')
]

failed = filtered_runs[filtered_runs.status == 'FAILED']
running = filtered_runs[filtered_runs.status == 'RUNNING']
finished = filtered_runs[filtered_runs.status == 'FINISHED']

print(len(failed))
print(len(running))
print(len(finished))

grouped_df = finished.groupby([
    'params.imgs_params/fig_params/linewidth',
    'params.imgs_params/fig_params/markersize',
    'params.imgs_params/input_type',
    'params.imgs_params/use_err',
    'params.training/lr',
])['metrics.f1/test'].agg(['mean', 'std']).reset_index()

grouped_df.rename(columns={'mean': 'metrics.f1/test_mean', 'std': 'metrics.f1/test_std'}, inplace=True)

print(f'Number experiments: {len(grouped_df)}')

grouped_df = grouped_df.sort_values(by=['metrics.f1/test_mean'], ascending=False)
dict_results = update_dict_results(grouped_df, dict_results, key='single_band', input_type='overlay')

grouped_df.head(10)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import PercentFormatter

fs = 16

# Etiquetas para las muestras por clase
spc_labels = ["20 samples p/class", "500 samples p/class", "Full dataset"]

# Configuración para las barras
x = np.arange(len(spc_labels))
width = 0.25

# Mejorar el diseño del gráfico con porcentajes y agrandar etiquetas y ticks
fig, ax = plt.subplots(figsize=(10, 5))  # Tamaño ajustado para mayor claridad

# Cambiar a colores más sobrios para un artículo científico
bar1 = ax.bar(
    x - width, 
    dict_results['single_band']['list_mean_by_spc'], 
    width, yerr=dict_results['single_band']['list_std_by_spc'], 
    label="Overlay (one-band)", 
    color="#4C72B0", 
    capsize=5
    )

bar2 = ax.bar(
    x, 
    dict_results['two_bands_overlay']['list_mean_by_spc'], 
    width, yerr=dict_results['two_bands_overlay']['list_std_by_spc'], 
    label="Overlay (two-bands)", 
    color="#55A868", 
    capsize=5
    )

bar3 = ax.bar(
    x + width, 
    dict_results['two_bands_grid']['list_mean_by_spc'], 
    width, yerr=dict_results['two_bands_grid']['list_std_by_spc'], 
    label="Grid (two-bands)", 
    color="#C44E52", 
    capsize=5
    )

# Etiquetas más limpias y claras
ax.set_ylabel(f"Test F1 Score (\%)", fontsize=fs + 4, labelpad=12)
ax.set_xlabel("Dataset Size", fontsize=fs + 4, labelpad=12)
ax.set_title("SwinV2's performance on different MACHO datasets", fontsize=fs + 6, weight='bold', pad=12)
#ax.set_title("Models' performance on different MACHO datasets", fontsize=fs + 4, weight='bold', pad=12)
ax.set_xticks(x)
ax.set_xticklabels(["20 samples per class", "500 samples per class", "20,894 objects (full dataset)"], fontsize=fs + 2)
ax.tick_params(axis='y', labelsize=fs+1)
ax.set_ylim(0.52, 0.96)
ax.legend(fontsize=fs + 1, loc="upper left")

ax.yaxis.set_major_formatter(PercentFormatter(1.0))

# Añadir los valores sobre las barras, alineados con las barras de error
def add_labels_with_error(bars, stds):
    for bar, std in zip(bars, stds):
        height = bar.get_height()
        if height > 0:  # Solo etiquetar las barras que tienen valores
            ax.text(bar.get_x() + bar.get_width() / 2, height + std + 0.015, f"{height*100:.1f}\%",
                    ha="center", va="bottom", fontsize=fs, color="black")

add_labels_with_error(bar1, dict_results['single_band']['list_std_by_spc'])
add_labels_with_error(bar2, dict_results['two_bands_overlay']['list_std_by_spc'])
add_labels_with_error(bar3, dict_results['two_bands_grid']['list_std_by_spc'])

# Mejorar el diseño del gráfico con cuadrícula más ligera
ax.grid(axis="y", linestyle="--", alpha=0.8)
plt.tight_layout()
#os.makedirs('./images', exist_ok=True)
#plt.savefig("./images/Model_performance_MACHO.pdf", format='pdf', bbox_inches='tight')
plt.show()
```
