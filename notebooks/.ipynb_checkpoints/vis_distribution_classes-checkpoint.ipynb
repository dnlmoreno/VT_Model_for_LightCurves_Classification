```python
import pandas as pd
import matplotlib.pyplot as plt
import os
#os.environ["PATH"] = "/home/users/dmoreno2016/.TinyTeX/bin/x86_64-linux:" + os.environ["PATH"]

plt.rcParams.update({
    "text.usetex": True,
    "pgf.texsystem": "pdflatex",
    "font.family": "serif",
    "font.serif": ["Palatino"],
})

os.environ["CUDA_VISIBLE_DEVICES"] = "-1"

%load_ext autoreload
%autoreload 2
```

```python
path_save_results = 'images'

CB_color_cycle = ['#377eb8', '#ff7f00', '#4daf4a',
                  '#f781bf', '#a65628', '#984ea3',
                  '#999999', '#e41a1c', '#dede00']

blue = '#377eb8'
orange = '#ff7f00'
```

```python
ATAT_partition = pd.read_parquet('data/lightcurves/elasticc_1/ATAT_partition/partitions_v1.parquet')
ATAT_partition = ATAT_partition[
    (ATAT_partition.fold.isin([0, -1]))
    ]

ATAT_partition
```

```python
plt.rcParams.keys()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# ParÃ¡metros y configuraciÃ³n
fs = 14
order_classes = [
    'CART', 'Iax', '91bg', 'Ia', 'Ib/c', 'II', 'SN-like/Other', 'SLSN', 'PISN', 'TDE', 'ILOT', 'KN',
    'M-dwarf Flare', 'uLens', 'Dwarf Novae', 'AGN', 'Delta Scuti', 'RR Lyrae', 'Cepheid', 'EB'
]

# Filtrar datos de entrenamiento y validaciÃ³n
train_data = ATAT_partition[ATAT_partition['subset'].str.contains('train')]
val_data = ATAT_partition[ATAT_partition['subset'].str.contains('val')]

# Contar objetos por clase
train_counts = train_data['class_name'].value_counts().reindex(order_classes, fill_value=0)
val_counts = val_data['class_name'].value_counts().reindex(order_classes, fill_value=0)

# Crear la figura y ejes principales
fig, ax = plt.subplots(figsize=(14, 6))
bar_width = 0.4
index = range(len(order_classes))

# Crear barras para entrenamiento y validaciÃ³n
bars_train = ax.bar(index, train_counts, width=bar_width, color=blue, label='Training', edgecolor='black')
bars_val = ax.bar([i + bar_width for i in index], val_counts, color=orange, width=bar_width, label='Validation', edgecolor='black')

# Etiquetas y configuraciones del grÃ¡fico principal
ax.set_xlabel('Class', fontsize=fs + 2, labelpad=12)
ax.set_ylabel('Number of objects', fontsize=fs + 2, labelpad=12)
ax.set_title('Distribution of objects by class in ELAsTiCC dataset', fontsize=fs + 4, pad=12)
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(order_classes, rotation=45, ha='right', fontsize=fs)
ax.tick_params(axis='x', labelsize=fs)
ax.tick_params(axis='y', labelsize=fs)
ax.legend(fontsize=fs)
ax.grid(axis='y', linestyle='--', alpha=0.7)

# Clases para el zoom
classes_zoom = ['ILOT', 'KN', 'M-dwarf Flare', 'uLens', 'Dwarf Novae']
zoom_indices = [order_classes.index(cls) for cls in classes_zoom]

# Crear el subgrÃ¡fico de zoom con posiciÃ³n manual usando bbox_to_anchor
ax_inset = inset_axes(ax, width=2.5+0.3, height=1.5+0.3, bbox_to_anchor=(0.79, 0.95), bbox_transform=ax.transAxes, borderpad=2)
#ax_inset = inset_axes(ax, width=2.5+0.3, height=1.5+0.3, bbox_to_anchor=(0.73, 0.95), bbox_transform=ax.transAxes, borderpad=2)
bars_train_zoom = ax_inset.bar([i for i in range(len(classes_zoom))],
                               [train_counts[cls] for cls in classes_zoom],
                               width=bar_width, edgecolor='black', color=blue)
bars_val_zoom = ax_inset.bar([i + bar_width for i in range(len(classes_zoom))],
                             [val_counts[cls] for cls in classes_zoom],
                             width=bar_width, edgecolor='black', color=orange)

# ConfiguraciÃ³n del subgrÃ¡fico
ax_inset.set_xticks([i + bar_width / 2 for i in range(len(classes_zoom))])
ax_inset.set_xticklabels(classes_zoom, rotation=45, ha='right', fontsize=fs - 2)
ax_inset.tick_params(axis='x', labelsize=fs - 2)
ax_inset.tick_params(axis='y', labelsize=fs - 2)
ax_inset.grid(axis='y', linestyle='--', alpha=0.7)

bottom_y = ax_inset.get_position().y0

# LÃ­nea desde el borde inferior izquierdo del subgrÃ¡fico
ax.plot([10.6, 9.8], 
        [140000, bottom_y], 
        color='black', linestyle='--', alpha=0.8, linewidth=0.6)

# LÃ­nea desde el borde inferior derecho del subgrÃ¡fico
ax.plot([15.4, 14.6], 
        [140000, bottom_y], 
        color='black', linestyle='--', alpha=0.8, linewidth=0.6)


# Ajuste final y mostrar el grÃ¡fico
plt.tight_layout()
os.makedirs(path_save_results, exist_ok=True)
plt.savefig(f"{path_save_results}/Distribution_ELAsTiCC.pdf", format='pdf', bbox_inches='tight')
plt.show()
```
