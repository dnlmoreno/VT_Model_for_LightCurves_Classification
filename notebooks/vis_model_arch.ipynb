```python
import matplotlib.colors as mcolors

# Define the hex colors
colors = {
    0: '#00FF7F', 
    1: '#7FFF00', 
    2: '#FF007F', 
    3: '#FF7F00', 
    4: '#007FFF', 
    5: '#7F00FF', 
}

# Convert hex to RGB and normalize to [0, 1]
rgb_values = {key: mcolors.hex2color(value) for key, value in colors.items()}

# Convert to 0-255 scale
rgb_255_values = {key: tuple(int(c * 255) for c in value) for key, value in rgb_values.items()}

# Display the results
import pandas as pd
df = pd.DataFrame.from_dict(rgb_255_values, orient='index', columns=['R', 'G', 'B'])

df
```
